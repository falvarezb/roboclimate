```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
london_df = pd.read_csv("london_weather_historical_data.csv")
london_df['parsed_dt'] = london_df['dt_iso'].apply(lambda x: x[:19])
london_df.set_index(pd.DatetimeIndex(london_df['parsed_dt']), inplace=True)
```

```python
plt.plot(london_df['temp']['2019-12-01':'2019-12-12'], label='2019')
# plt.plot(london_df['temp']['2018-12-01':'2018-12-12'], label='2018')
plt.title("London temperature")
plt.legend()
plt.show()
```
